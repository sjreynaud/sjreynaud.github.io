```python
#4 Summary statistics / Summarize Dataset with Descriptive Statistics
df.describe(include='all')
```

**Step#4:** This step produces a comprehensive statistical overview of the dataset using df.describe(include='all'), which includes both numeric and non-numeric columns.
```python
#5 Missing values / Detect and Report Missing Values
missing = df.isnull().sum()
print("Missing values:\n", missing[missing > 0])
```

**Step#5:** This step scans the dataset for missing values by using df.isnull().sum() to count null entries in each column. It then filters the results to display only columns with at least one missing value, highlighting potential data quality issues. The output reveals that several columnsâ€”including R_vs_NR, dialysis_date, death_date, and multiple biopsy-related fieldsâ€”have substantial gaps, with some missing all 579 entries. This diagnostic is essential for guiding imputation, exclusion, or further investigation before modeling, ensuring that missingness is addressed systematically.

```python
#6 Value counts for categorical columns / Analyze Frequency of Categorical Variables
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n{col} value counts:\n", df[col].value_counts())
```

**Step#6:** This step identifies all categorical columns in the datasetâ€”those with data type objectâ€”and prints the frequency of each unique value within them. By looping through each categorical column and applying value_counts(), it reveals the distribution of responses for variables like treatment_response, gender_0_1, immunosuppresion_y_n, and acearb_y_n. This is crucial for understanding class balance, spotting dominant categories, and preparing for encoding or stratified analysis in downstream modeling.

```python
#7 Correlation matrix / Visualize Feature Correlations with Heatmap.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Step#7:** This step generates a correlation matrix heatmap to explore linear relationships among numeric features in the dataset. Using df.corr(numeric_only=True), it computes Pearson correlation coefficients between all numeric columns, which range from -1 (perfect negative) to +1 (perfect positive). The heatmap, rendered with Seaborn and Matplotlib, uses a coolwarm color palette to visually encode these relationshipsâ€”red for strong positive, blue for strong negative, and white for neutral. Annotated values inside each cell make it easy to spot highly correlated pairs, such as between serum_creatinine_mgdl and yearstodialysis, which may inform feature selection, multicollinearity checks, or clinical interpretation
